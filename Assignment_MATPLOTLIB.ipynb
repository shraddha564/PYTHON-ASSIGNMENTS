```python
**See if you can resize the plot by adding the figsize()argument in plt.subplots() are copying and pasting your previous code.**
```

```python
fig,axes=plt.subplots(nrows=1, ncols=2)

a1=fig,axes=plt.subplots(figsize=(6,2))
for v in a1:
    v.plot(x,y,'r')
    v.set_xlabel('x')
    v.set_ylabel('y')
    v.set_title('title')
a2=fig,axes=plt.subplots(figsize=(6,2))
for v in a2:
    v.plot(x,y,'r')
    v.set_xlabel('x')
    v.set_ylabel('y')
    v.set_title('title')
```
